```python
import requests
import bs4
import pprint as pp
from bs4 import BeautifulSoup
```

```python
base_url = 'https://kinoteatr.ru/'
films_url = 'kinoafisha/'
data = input()
dataAll = '/?date=' + str(data)
r = requests.get(f'{base_url}{films_url}{dataAll}')
soap = BeautifulSoup(r.text, 'html');
j = 0;
print("{:^150}".format(data))
for i in soap.findAll('div', class_ = 'col-md-2 col-sm-6 col-xs-12 movie_card'):
    name = i.find('span', class_ = 'movie_card_header title').text.strip()
    date = i.find('div',  class_ = 'col col-md-3 col-sm-4 col-xs-12 ');
    date = "/date=2020-12-01"
    r2 = requests.get(f'{base_url}{films_url}{date}')
    soapDate = BeautifulSoup(r2.text, 'html');
    cinemasLink = i.find('a', class_ = 'movie_card_clickable_zone gtm-ec-list-item-movie').get('href');
    r1 = requests.get(f'{cinemasLink}');
    soapLink = BeautifulSoup(r1.text, 'html');
    j += 1;
    cinemas = []
    for c in soapLink.findAll('div', class_='cinema'):
        text=c.find('p', class_='text').text.strip();
        namec=c.find('div', class_='name').text.strip()
        timelst=[]
        for g in c.findAll('div', class_='time'):
            time=g.text.strip()[0:5]
            timelst.append(time);
        print(name, namec, text, timelst)
```
